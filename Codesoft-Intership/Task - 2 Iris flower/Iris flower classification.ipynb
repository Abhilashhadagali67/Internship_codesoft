# Iris Flower classification:

# Importing Modules:

```python
import pandas as pd
```

```python
df = pd.read_csv('IRIS.csv')
df
```

```python
df.isnull().sum()
```

# Text Preprocessing:

```python
#Flower name has string format. So convert it into int format:

from sklearn.preprocessing import LabelEncoder
le_gender = LabelEncoder()
```

```python
df['flower_name'] = le_gender.fit_transform(df['species'])
df
```

```python
##0 for Iris-setosa 1 for Iris-Versicolor and 2 for Iris-virginica

df.drop(['species'],axis = 'columns',inplace = True)
df
```

```python
#Making a seperate coulmn.

y=df.drop(['sepal_length','sepal_width','petal_length','petal_width'],axis = 'columns')
```

```python
df.drop(['flower_name'],axis = 'columns',inplace = True)
df
```

# Model Building:

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df,y,test_size=0.2)

def functionmodel(model):
    model.fit(x_train,y_train)
    print("Accuracy: ",model.score(x_test,y_test))
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
functionmodel(model)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
functionmodel(model)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
functionmodel(model)
```
